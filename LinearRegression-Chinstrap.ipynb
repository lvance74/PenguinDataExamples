```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

penguins = sns.load_dataset('penguins')
```

```python
penguins = penguins.dropna()
penguins.info()
```

```python
chinstrap = penguins[penguins["species"]=="Chinstrap"]
display(chinstrap)
```

```python
x = chinstrap.iloc[:, 3:4]
display(x)

y = chinstrap.iloc[:, 2:3]
display(y)
```

```python
lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x)
```

```python
plt.scatter(chinstrap['bill_depth_mm'], chinstrap['bill_length_mm'], color = 'blue')
plt.plot(x,y_pred, color='red')
plt.title('Bill Depth vs Bill Length for Chinstrap')
plt.xlabel('Bill Depth')
plt.ylabel('Bill Length')
```

```python
i = lr.intercept_
print("Intercept", i)

c = lr.coef_
print("Coefficient",c)

s = lr.score(x,y)
print(s)
```
