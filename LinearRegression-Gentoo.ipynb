```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

penguins = sns.load_dataset('penguins')
```

```python
penguins = penguins.dropna()
penguins.info()
```

```python
gentoo = penguins[penguins["species"]=="Gentoo"]
display(gentoo)
```

```python
x = gentoo.iloc[:, 3:4]
display(x)

y = gentoo.iloc[:, 2:3]
display(y)
```

```python
lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x)
```

```python
plt.scatter(gentoo['bill_depth_mm'], gentoo['bill_length_mm'], color = 'blue')
plt.plot(x,y_pred, color='red')
plt.title('Bill Depth vs Bill Length for Gentoo')
plt.xlabel('Bill Depth')
plt.ylabel('Bill Length')
```

```python
c = lr.coef_
print("Coefficient",c)

i = lr.intercept_
print("Intercept", i)

s = lr.score(x,y)
print(s)
```

```python
test_x = 13.2
result = c * test_x + i
print(result)
```
