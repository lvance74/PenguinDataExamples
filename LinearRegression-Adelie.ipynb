```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

penguins = sns.load_dataset('penguins')
```

```python
penguins = penguins.dropna()
penguins.info()
```

```python
adelie = penguins[penguins["species"]=="Adelie"]
display(adelie)
```

```python
adelie.corr(numeric_only=True)
```

```python
x = adelie.iloc[:, 3:4]
display(x)

y = adelie.iloc[:, 5:6]
display(y)
```

```python
lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x)
```

```python
plt.scatter(adelie['bill_depth_mm'], adelie['body_mass_g'], color = 'blue')
plt.plot(x,y_pred, color='red')
plt.title('Bill Depth vs Body Mass for Adelie')
plt.xlabel('Bill Depth')
plt.ylabel('Body Mass')
```

```python
i = lr.intercept_
print("Intercept", i)

c = lr.coef_
print("Coefficient",c)

s = lr.score(x,y)
print(s)
```
