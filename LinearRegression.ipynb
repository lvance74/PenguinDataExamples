```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

penguins = sns.load_dataset('penguins')
```

```python
display(penguins)
```

```python
penguins.info()
```

```python
penguins = penguins.dropna()
penguins.info()
```

```python
penguins.describe()
```

```python
x = penguins.iloc[:, 3:4]
display(x)

y = penguins.iloc[:, 2:3]
display(y)
```

```python
lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x)
```

```python
plt.scatter(penguins['bill_depth_mm'], penguins['bill_length_mm'], color = 'blue')
plt.plot(x,y_pred, color='red')
plt.title('Bill Depth vs Bill Length for Gentoo')
plt.xlabel('Bill Depth')
plt.ylabel('Bill Length')
```

```python
i = lr.intercept_
print("Intercept", i)

c = lr.coef_
print("Coefficient",c)

s = lr.score(x,y)
print(s)
```
